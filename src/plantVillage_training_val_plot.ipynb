```python
import matplotlib.pyplot as plt

# Data for 12 epochs
epochs = list(range(1, 13))
train_loss = [0.7702, 0.0997, 0.0543, 0.0347, 0.0242, 0.0172, 0.0128, 0.0106, 0.0083, 0.0076, 0.0083, 0.0084]
val_loss   = [0.0526, 0.0241, 0.0189, 0.0194, 0.0109, 0.0125, 0.0105, 0.0082, 0.0079, 0.0078, 0.0081, 0.0077]
train_acc = [82.19, 97.71, 98.68, 99.09, 99.37, 99.61, 99.70, 99.75, 99.83, 99.85, 99.81, 99.79]
val_acc   = [98.69, 99.40, 99.52, 99.42, 99.72, 99.69, 99.74, 99.80, 99.84, 99.83, 99.85, 99.83]

# Print training logs with best model saving
best_val_acc = 0
for i, ep in enumerate(epochs):
    tl = train_loss[i]
    ta = train_acc[i]
    vl = val_loss[i]
    va = val_acc[i]
    print(f"Epoch [{ep}/12] - Train Loss: {tl:.4f}, Train Acc: {ta:.2f}% | Val Loss: {vl:.4f}, Val Acc: {va:.2f}%")


# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Loss subplot
ax1.plot(epochs, train_loss, marker='o', label='Train Loss')
ax1.plot(epochs, val_loss, marker='s', label='Validation Loss')
ax1.set_title('Training and Validation Loss')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.set_xticks(epochs)
ax1.legend()
ax1.grid(True)

# Accuracy subplot
ax2.plot(epochs, train_acc, marker='o', label='Train Accuracy')
ax2.plot(epochs, val_acc, marker='s', label='Validation Accuracy')
ax2.set_title('Training and Validation Accuracy')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy (%)')
ax2.set_xticks(epochs)
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
